```python
#openig the csv file of syria as a test file
import csv, os, shutil
csvfile = open('M:\masters\semester 4\web scraping\output files\TopicStats\Syria.csv', newline='') 
spamreader = csv.reader(csvfile, delimiter=',')
MpNamesFolder = [] #a list with all the mp names in the Syria.csv from the 2016_2015 statistics
SubjectName = [] #a list with the names that should be present in the mp folders from the downloaded data
directory = 'M:\masters\semester 4\web scraping\output files\TopicData\\2016_2015' #directory of the data      
copydirectory = 'M:\masters\semester 4\PythonTrials\Syria_speeches'
```

```python
#read every line in the csv and extract the name and save it in MpNamesFolder, create the file name and append it to the SubjectName
#List, Every item in the MpNamesFolder corresponds to the Items in SubjectName list.
for row in spamreader:
    MpNamesFolder.append(row[0])
    if not row[2]:
        SubjectNamePDate = "Syria"+".txt"
        SubjectName.append(SubjectNamePDate)
        #print(SubjectNamePDate)
    else:
        SubjectDate = row[2].replace(" ", "_")
        SubjectNamePDate = "Syria"+'_'+SubjectDate+".txt"
        SubjectName.append(SubjectNamePDate)
        #print(SubjectNamePDate)
        #print(row[0])
        #print(', '.join(row))
```

```python
folderDirectory = ""
fileDirectory = ""
counter = 0
for name in MpNamesFolder:
    filename = SubjectName[counter]
    foldername = name.strip()
    folderDirectory = os.path.join(directory,foldername)
    counter = counter + 1
    fileDirectory = os.path.join(folderDirectory, filename)
    folderNameClean = foldername.replace(" ", "_")
    newFileName = folderNameClean+"_"+filename
    print(fileDirectory)
    if (os.path.isfile(fileDirectory)):
        shutil.copy(fileDirectory, copydirectory)
        newDstFileName = os.path.join(copydirectory, newFileName)
        os.rename(fileDirectory, newDstFileName)
        print("done copying")
```
